# Visualization of benchmark results

## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load CSV

```python
fp16 = pd.read_csv("../results/phi2_fp16_baseline.csv")
int8 = pd.read_csv("../results/phi2_int8.csv")
int4 = pd.read_csv("../results/phi2_int4.csv")
```

```python
df = pd.concat([fp16, int8, int4])

print(df)
```

## 1. RAM Usage

In this plot is shown the comprisation of RAM usage between the precisions of Phi-2.

```python
plt.figure(figsize=(8,5))

plt.bar(
    df["precision"],
    df["ram_usage_gb"]
)

plt.xlabel("Precision")
plt.ylabel("RAM Usage (GB)")
plt.title("RAM Usage by Quantization Level")

plt.savefig("../figures/phi-2_ram_usage.png", bbox_inches="tight")

plt.show()
```

## 2. Tokens/SEC

In this second plot is shown the results for tokens per second that the models produce

```python
plt.figure(figsize=(8,5))

plt.bar(
    df["precision"],
    df["tokens_per_second"]
)

plt.xlabel("Precision")
plt.ylabel("Tokens per Second")
plt.title("Inference Throughput")

plt.savefig("../figures/phi-2_tokens_per_sec.png", bbox_inches="tight")

plt.show()
```

## 3. Inference time

```python
plt.figure(figsize=(8,5))

plt.bar(
    df["precision"],
    df["inference_time"]
)

plt.xlabel("Precision")
plt.ylabel("Inference Time (sec)")
plt.title("Inference Latency")

plt.savefig("../figures/phi-2_inference.png", bbox_inches="tight")

plt.show()
```
